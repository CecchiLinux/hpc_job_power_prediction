```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def autolabel(rects, axs):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        #print(height)
        if(height > 0.0):
            axs.text(rect.get_x() + rect.get_width()/2., height,
                '%.2f' % height,
                ha='center', va='bottom', fontsize = 10)
```

```python
def my_hist_plot_2(x_values, slowdown_values, waittime_values, y_label, y_lim):

    N = len(x_values)

    ind = np.arange(N)  # the x locations for the groups
    width = 0.28      # the width of the bars

    fig, ax = plt.subplots(figsize=(8, 4))
    rects1 = ax.bar(ind, slowdown_values, width, color='b')
    rects2 = ax.bar(ind+width, waittime_values, width, color='orange')

    # add some text for labels, title and axes ticks
    ax.set_ylabel(y_label, fontsize=14)
    ax.set_xlabel('power budget (% of initial power)', fontsize=14)
    ax.set_xticks(ind + width/2)
    ax.set_xticklabels(x_values)

    axes = plt.gca()
    axes.set_ylim([0, y_lim])

    ax.legend((rects1[0], rects2[0]), ('slowdown', 'waiting time'), fontsize=14)

    autolabel(rects1, ax)
    autolabel(rects2, ax)

    plt.xticks(size = 12)
    plt.yticks(size = 12)
    plt.show()
```

```python
def my_hist_plot(x_values, y_values, y_label, y_lim, color='b'):

    N = len(x_values)

    ind = np.arange(N)  # the x locations for the groups
    width = 0.20      # the width of the bars

    fig, ax = plt.subplots(figsize=(6, 4))
    rects1 = ax.bar(ind, y_values, width, color=color)

    # add some text for labels, title and axes ticks
    ax.set_ylabel(y_label, fontsize=14)
    ax.set_xlabel('power budget (% of initial power)', fontsize=14)
    ax.set_xticks(ind)
    ax.set_xticklabels(x_values)

    axes = plt.gca()
    axes.set_ylim([0, y_lim])

    #ax.legend((rects1[0], rects2[0], rects3[0], rects4[0], rects5[0]), 
    #('perNode', 'perUser', 'perUserNode', 'perUserJobname', 'perUserJobnameNode'))

    autolabel(rects1, ax)

    plt.xticks(size = 12)
    plt.yticks(size = 12)
    plt.show()
```

```python
# 10 30 50 70 90 100
avg_waittimes = [0, 2293.6/399.42, 605.81/399.42, 419.51/399.42, 400.31/399.42, 399.42/399.42]
slowdowns = [0, 27.91/1.77, 2.55/1.77, 2.68/1.77, 2.66/1.77, 1.77/1.77]
xticks = ['10', '30', '50', '70', '90', '100']
my_hist_plot(xticks, slowdowns, 'slowdown', 20, 'b')
my_hist_plot(xticks, avg_waittimes, 'avg. waiting time (ms)', 7, 'r')
```
