```python
import pandas as pd
import numpy as np

'''
Author: Enrico Ceccolini

    job's duration longer than 1 mintute version
    
    remove from the original data jobs with:
    - missing end_time
    - missing shedule informations (in which nodes was computed)
    - wrong schedule informations (scheduled on unexisting node)
    
    output the csv of the cleaned data
'''

datadir = "/datasets/eurora_data/db1/"
infile_jobs_original = datadir + "jobs"
outfile_jobs_cleaned = datadir + "jobs_cleaned"

suffix = "_1min_"
### select an interval from
## 1 settings wholeData
interval_comment_whole = "WholeData"

### select an interval from
## 2 settings Andrea
#interval_comment = "Andrea"
#start_time = pd.to_datetime('2014-03-31')
#end_time = pd.to_datetime('2014-05-01')

## 3 settings Alina
interval_comment = "Alina"
start_time = pd.to_datetime('2014-07-30')
end_time = pd.to_datetime('2014-11-01')

# read dataset
jobs_original = pd.read_csv(infile_jobs_original + ".csv", index_col=0) # Read a comma-separated values (csv) file into DataFrame
jobs_original.shape
```

```python
jobs_original.head(1)
```

```python
# remove the jobs with missing end-time
jobs_completed = jobs_original[jobs_original.end_time != '0000-00-00 00:00:00']
print("completed_jobs_data contains {} records".format(jobs_completed.shape[0]))
```

```python
interval_jobs = jobs_completed[pd.to_datetime(jobs_completed['run_start_time']) >= start_time]
interval_jobs = interval_jobs[pd.to_datetime(interval_jobs['end_time']) <= end_time]
print("andrea's interval contains {} records".format(interval_jobs.shape[0]))
```

```python
long_jobs = interval_jobs[pd.to_datetime(interval_jobs['end_time']) - pd.to_datetime(interval_jobs['run_start_time']) > np.timedelta64(1, 'm')]
print("{}'s interval contains {} long jobs".format(interval_comment, long_jobs.shape[0]))
```

recover the missing required resources

```python
infile_jobs_to_nodes = datadir + "job_nodes.csv"

jobs_to_nodes_whole_data = pd.read_csv(infile_jobs_to_nodes, index_col=0)
print("jobs_to_nodes_whole_data contains {} records".format(jobs_to_nodes_whole_data.shape[0]))

### clean the data
# remove jobs runned on the inexistent node 129
jobs_to_nodes_whole_data = jobs_to_nodes_whole_data[jobs_to_nodes_whole_data['node_id'] != 129] 
# remove jobs with the same id that runned on the same node
jobs_to_nodes_whole_data = jobs_to_nodes_whole_data.drop_duplicates(subset=['job_id_string', 'node_id'])
print("after the clean, jobs_to_nodes_whole_data contains {} records".format(jobs_to_nodes_whole_data.shape[0]))

merged_jobs_to_nodes = pd.merge(jobs_to_nodes_whole_data, long_jobs, on='job_id_string')
print("merged dataframe rows: {}".format(merged_jobs_to_nodes.shape[0]))
merged_jobs_to_nodes.head(1)
```

```python
i = 0
for index, row in long_jobs.iterrows():
    print("{}/{}".format(i, long_jobs.shape[0]))
    i += 1
    if (row['node_req'] == 0):
        job_to_nodes = merged_jobs_to_nodes[merged_jobs_to_nodes['job_id_string'] == row['job_id_string']]
        nnodes = job_to_nodes.shape[0]
        ncpus = job_to_nodes['ncpus'].sum()
        ngpus = job_to_nodes['ngpus'].sum()
        nmics = job_to_nodes['nmics'].sum()
        mem_requested = job_to_nodes['mem_requested'].sum()

        #print(large_jobs.loc[index])
        #print(ncpus)
        long_jobs.at[index, 'node_req'] = nnodes
        long_jobs.at[index, 'cpu_req'] = ncpus
        long_jobs.at[index, 'gpu_req'] = ngpus
        long_jobs.at[index, 'mic_req'] = nmics
        long_jobs.at[index, 'mem_req'] = mem_requested
```

```python
long_jobs = long_jobs[long_jobs['node_req'] != 0]
print("number of the jobs after the clean: {}".format(long_jobs.shape[0]))
```

```python
out_file = datadir + "CPUs/" + interval_comment + "/" + interval_comment +"_long_jobs"
long_jobs.to_csv(out_file + ".csv")
```
